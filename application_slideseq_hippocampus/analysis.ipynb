```python
import pandas as pd
import anndata as ad
import numpy as np
import scanpy as sc
from plotnine import *
```

```python
coords = pd.read_csv("Puck_200306_03_coords.csv")
```

```python
coords
```

```python
expression = pd.read_csv("Puck_200306_03.digital_expression.txt", sep="\t")
expression.index = expression["GENE"]
expression.drop(columns=["GENE"], inplace=True)
expression = expression.T
```

```python
adata = ad.AnnData(expression)
adata.obs["x"] = coords["xcoord"].values
adata.obs["y"] = coords["ycoord"].values
```

```python
adata.write_h5ad("somatosensory_cortex.h5ad")
```

```python
adata
```

```python
plot = (
    ggplot(adata.obs, aes(x="x", y="y"))
    + geom_point(size=0.01)
)
plot.show()
```
