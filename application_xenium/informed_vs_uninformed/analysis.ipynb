```python
from prismo import PRISMO
import matplotlib.pyplot as plt
import numpy as np
from plotnine import *
import pandas as pd
```

```python
models_informed = []
models_uninformed = []

for i in range(10):
    models_informed.append(PRISMO.load(f"models/informed_{i}.h5"))
    models_uninformed.append(PRISMO.load(f"models/uninformed_{i}.h5"))
```

```python
z_informed = models_informed[0].get_factors()["group_xenium"]
z_uninformed = models_uninformed[0].get_factors()["group_xenium"]
```

```python
corrs = np.zeros((z_informed.shape[1], z_uninformed.shape[1]))
for i in range(z_informed.shape[1]):
    for j in range(z_uninformed.shape[1]):
        corrs[i, j] = np.corrcoef(z_informed.iloc[:, i], z_uninformed.iloc[:, j])[0, 1]
```

```python
plt.imshow(corrs, cmap="coolwarm")
```
